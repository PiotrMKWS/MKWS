```python
import numpy as np
import matplotlib.pyplot as plt
import cantera as ct
%matplotlib inline

n_rev = 3000./60.
bore = 0.084 # [m]
stroke = 0.088 # [m]
piston_area = bore * np.pi ** 2 * 0.25 # [m^2]
cmpr_ratio = 19. # [-]
disp_vol = stroke * bore ** 2 * np.pi * 0.25 # [m^3]
print(disp_vol)
cmpr_vol = disp_vol/(cmpr_ratio-1) # [m^3]
print(cmpr_vol)
print(disp_vol/cmpr_vol)
lambda_ratio = 0.1 # crank radius divided by connecting rod length
crank_radius = stroke/2.0
n_rad = 2 * np.pi * n_rev

def crank_angle(t):
    return np.remainder(2 * np.pi * n_rev * t, 2 * np.pi)

def crank_angle_deg(t):
    return crank_angle(t) * 180/np.pi

def piston_speed(t):
    return - crank_radius * n_rad * (np.sin(crank_angle(t))+0.5*lambda_ratio*np.sin(crank_angle(t)))

def mfc_mdot(t):
    if (crank_angle_deg(t) > 170 and crank_angle_deg(t) < 190):
        return 0.05
    else:
        return 0.0
    
def injector_mdot(t):
    injector_open = 170. / 180. * np.pi
    injector_close = 190. / 180. * np.pi
    injector_mass = 1.0e-3  # kg
    if (crank_angle_deg(t) > 170 and crank_angle_deg(t) < 190):
        return injector_mass / ((injector_close - injector_open) / (2. * np.pi * n_rev))
    else:
        return 0.0

def k_v_exh(t):
    if (crank_angle_deg(t) > 190 and crank_angle_deg(t) < 360):
        return 1
    else:
        return 0.0
    
def k_v_inl(t):
    if (crank_angle_deg(t) > 0 and crank_angle_deg(t) < 170):
        return 1
    else:
        return 0.0    
    

air = ct.Solution('gri30.xml')
spark = ct.Solution('gri30.xml')
fuel = ct.Solution('gri30.xml')

air.TPX = 300, ct.one_atm, "O2:2.0, N2:7.52"
spark.TPX = 8000, ct.one_atm, "H:1"
fuel.TPX = 300, ct.one_atm, "CH4:1.0"

r1 = ct.Reactor(air)
env = ct.Reservoir(air)
r_sp = ct.Reservoir(spark)
fueltank = ct.Reservoir(fuel)
inlet = ct.Reservoir(air)
outlet = ct.Reservoir(air)

igniter = ct.MassFlowController(r_sp,r1,mdot = mfc_mdot)
injector = ct.MassFlowController(fueltank,r1,mdot = injector_mdot) 
exhaust = ct.Valve(r1, outlet, K=k_v_exh)
inlet = ct.Valve(inlet, r1, K=k_v_inl)
piston = ct.Wall(r1, env, velocity = piston_speed)

r1.volume = 0.1

times = np.zeros(2000)
angle = np.zeros_like(times)
speed = np.zeros_like(times)
r1_volume = np.zeros_like(times)
r1_t = np.zeros_like(times)
r1_p = np.zeros_like(times)
r1_H = np.zeros_like(times)
angle_deg = np.zeros_like(times)
sim = ct.ReactorNet([r1])

for n in range (2000):
    t_i = n*1e-6        
    times[n] = t_i
    angle[n] = crank_angle(t_i)
    angle_deg[n] = crank_angle_deg(t_i)
    speed[n] = piston_speed(t_i)
    r1_volume[n] = r1.volume
    r1_t[n] = r1.T
    r1_p[n] = 1e-5*r1.thermo.P
    r1_H[n] = r1.thermo['H'].X
    print('%10.1f, %10.1f, %10.1f %10.8f, %10.3f, %10.3f' % (n, r1.T, r1.thermo.P, r1.thermo['H'].X, r1.thermo['CH4'].X, crank_angle_deg(t_i)))
    sim.advance(t_i)

plt.plot(times, angle)
plt.xlabel('time [s]')
plt.ylabel('angle [rad]')
plt.show()

plt.plot(times, speed)
plt.xlabel('time [s]')
plt.ylabel('speed [m/s]')
plt.show()

plt.plot(times, r1_volume)
plt.xlabel('time [s]')
plt.ylabel('vol [m^3]')
plt.show()

plt.plot(times, r1_t)
plt.xlabel('time [s]')
plt.ylabel('temp')
plt.show()

plt.plot(times, r1_p)
plt.xlabel('time [s]')
plt.ylabel('press [bar]')
plt.show()

plt.plot(times, r1_H)
plt.xlabel('time [s]')
plt.ylabel('r1_H')
plt.show()

plt.plot(times, angle_deg)
plt.xlabel('time [s]')
plt.ylabel('angle [deg]')
plt.show()
```

```python
#Source - SDToolbox demo named demo_CJstate

from SDToolbox import *

#Letting user specify the fuel for stoichometric calculations
mixture = raw_input("Choose fuel (methane/ethane/propane): ")

if mixture == "methane":
    P1=one_atm; P1atm=P1/one_atm;
    T1=300;
    q='CH4:1 O2:2 N2: 7.52';
    mech = 'gri30_highT.cti';
elif mixture == "ethane":
    P1=one_atm; P1atm=P1/one_atm;
    T1=300;
    q='C2H6:1 O2:3.5 N2: 13.6';
    mech = 'gri30_highT.cti';
elif mixture == "propane":
    P1=one_atm; P1atm=P1/one_atm;
    T1=300;
    q='C3H8:1 O2:5 N2:18.8';
    mech = 'gri30_highT.cti';
else:
    print("Specified mixture not available. Write 'methane', 'ethane' or 'propane'.")

#Calculating the Chapman Jouget detonation speed using least squares approximation method
#Last parameter in CJspeed() function is not a valid input
[cj_speed,R2] = CJspeed(P1, T1, q, mech, 0);

#Calculating the Chapman Jouget detonation post-shock state
gas = PostShock_eq(cj_speed, P1, T1, q, mech)
Ps = gas.P/one_atm

#Calculating the Chapman Jouget detonation frozen and equilibrium sound speeds
#For equilibrium sounds speed, entropy and specific volume are held constant
[a_equil,a_frozen] = equilSoundSpeeds(gas)

#Exporting CJ data: frozen/equilibrium speed, temperature, pressure to a file in current directory
output = open('results_' + mixture + '.txt', 'w')

output.write('For ' + q + ' with P1 = %.2f atm & T1 = %.2f K using ' % (P1atm,T1) + mech +'\n')
output.write('CJ Speed is %.2f m/s' % cj_speed +'\n')
output.write('The CJ State is %.2f atm & %.2f K' % (Ps,gas.T)+ '\n')
output.write('The sound speeds are: a_frozen = %.2f m/s & a_equilibrium = %.2f m/s' % (a_frozen,a_equil))

output.close()

#Printing the data in the terminal for good measure
print ' '
print 'CJ State'
gas()
Ps = gas.P/one_atm

print ' '
print 'For ' + q + ' with P1 = %.2f atm & T1 = %.2f K using ' % (P1atm,T1) + mech 
print 'CJ Speed is %.2f m/s' % cj_speed

print 'The CJ State is %.2f atm & %.2f K' % (Ps,gas.T)

[a_equil,a_frozen] = equilSoundSpeeds(gas)

print 'The sound speeds are: a_frozen = %.2f m/s & a_equilibrium = %.2f m/s' % (a_frozen,a_equil)
print ' '
```
